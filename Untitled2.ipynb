```python
import importlib

packages = [
    "rasterio", "geopandas", "azure.storage.blob", "sklearn",
    "matplotlib", "scipy", "pandas", "numpy", "shapely",
    "pystac_client", "requests", "boto3"
]

for pkg in packages:
    try:
        mod = importlib.import_module(pkg)
        version = getattr(mod, "__version__", "unknown version")
        print(f"✅ {pkg} ({version})")
    except ImportError:
        print(f"❌ {pkg} — not installed")
```
